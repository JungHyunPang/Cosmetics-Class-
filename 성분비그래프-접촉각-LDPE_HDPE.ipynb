```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
gap = 0.01
xmin, xmax, ymin, ymax, zmin, zmax = 0, 2, 0, 2, 0, 2
```

```python
xH = []
yH = []
zH = []
fH = []

xL = []
yL = []
zL = []
fL = []

alpha = 74.
beta = 38.2
gamma = 115.
for i in range(101):
    x = gap * i
    for j in range(101):
        for k in range(101):
            if i+j+k == 100:                
                y = gap*j
                z = 1. - x - y
                f = alpha * x + beta * y + gamma*z
                if (f>=60. and f<=90.) :
                    xH.append(x)
                    yH.append(y)
                    zH.append(z)
                    fH.append(f)
                elif f>90. :
                    xL.append(x)
                    yL.append(y)
                    zL.append(z)
                    fL.append(f)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xL, yL, zL, c=fL, marker='o', s=15, cmap='Reds')
ax.scatter(xH, yH, zH, c=fH, marker='o', s=15, cmap='Blues')
ax.view_init(30, 30)

plt.show()
```
