```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
gap = 0.01
xmin, xmax, ymin, ymax, zmin, zmax = 0, 2, 0, 2, 0, 2
```

```python
xs = []
ys = []
zs = []
fs = []
alpha = 74.
beta = 38.2
gamma = 115.
for i in range(101):
    x = gap * i
    for j in range(101):
        for k in range(101):
            if i+j+k == 100:                
                y = gap*j
                z = 1. - x - y
                f = alpha * x + beta * y + gamma*z
                xs.append(x)
                ys.append(y)
                zs.append(z)
                fs.append(f)
```

```python
cmin = min(fs)
cmax = max(fs)
print(cmin, cmax)
```

```python
fig = plt.figure(figsize=(6, 6))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(xs, ys, zs, c=fs, marker='o', s=15, cmap='Greens')
ax.view_init(30, 30)

plt.show()
plt.colorbar()
```
